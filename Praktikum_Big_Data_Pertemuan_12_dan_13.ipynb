# Pertemuan 12: Data Cleaning and Preparation using Pandas

### Exercise 1

```python
# Exercise 1: Identifying and Handling Missing Data
import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age': [24, 30, None, 22, 35],
    'Salary': [48000, None, 57000, None, 60000]
}
df = pd.DataFrame(data)

# Filling missing values and dropping rows
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df.dropna(subset=['Name'], inplace=True)
print('After cleaning:\n', df)
```

## Use Matplotlib to create a multi-bar chart to compare data across groups.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_df = pd.read_csv(url)

# Grouping data by 'Pclass' (Passenger Class) and 'Survived'
grouped_data = titanic_df.groupby(['Pclass', 'Survived'])['PassengerId'].count().reset_index()

# Creating a multi-bar chart
fig, ax = plt.subplots(figsize=(8, 6))

# Width of each bar
bar_width = 0.35

# Positions of bars on x-axis
pclass_positions = grouped_data['Pclass'].unique()
bar_positions_survived = pclass_positions
bar_positions_not_survived = pclass_positions + bar_width

# Plotting bars for survived and not survived passengers
survived_data = grouped_data[grouped_data['Survived'] == 1]['PassengerId']
not_survived_data = grouped_data[grouped_data['Survived'] == 0]['PassengerId']

ax.bar(bar_positions_survived, survived_data, width=bar_width, label='Survived')
ax.bar(bar_positions_not_survived, not_survived_data, width=bar_width, label='Not Survived')

# Setting chart labels and title
ax.set_xlabel('Passenger Class')
ax.set_ylabel('Number of Passengers')
ax.set_title('Survival Rate by Passenger Class')
ax.set_xticks(pclass_positions + bar_width / 2) # Centering x-axis ticks
ax.set_xticklabels(pclass_positions)
ax.legend()

plt.show()
```

## Generate scatter plots with Seaborn to explore relationships between variables in a dataset.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_df = pd.read_csv(url)

# Mengisi missing values sebelum membuat scatter plot
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)

# Scatter plot antara 'Age' dan 'Fare'
sns.scatterplot(x='Age', y='Fare', data=titanic_df, hue='Survived')
plt.title('Relationship between Age, Fare, and Survival')
plt.show()

# Scatter plot antara 'Fare' dan 'Pclass'
sns.scatterplot(x='Fare', y='Pclass', data=titanic_df, hue='Survived', palette='viridis')
plt.title('Relationship between Fare, Pclass, and Survival')
plt.show()
```

## Customize plot layouts and color schemes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_df = pd.read_csv(url)

# Mengisi missing values
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)

# Kustomisasi layout dan skema warna
sns.set_style("whitegrid")  # Mengatur style plot

# Scatter plot 1: Age vs. Fare
plt.figure(figsize=(10, 6))  # Mengatur ukuran figure
sns.scatterplot(x='Age', y='Fare', data=titanic_df, hue='Survived',
                palette="magma",  # Mengatur palet warna
                size="Pclass", sizes=(50, 200),  # Mengatur ukuran titik berdasarkan Pclass
                alpha=0.7)  # Mengatur transparansi titik
plt.title('Relationship between Age, Fare, and Survival', fontsize=16)  # Mengatur judul dan ukuran font
plt.xlabel('Age', fontsize=12)  # Mengatur label sumbu x dan ukuran font
plt.ylabel('Fare', fontsize=12)  # Mengatur label sumbu y dan ukuran font
plt.legend(title="Survived", loc="upper right")  # Mengatur posisi dan judul legend
plt.show()

# Scatter plot 2: Fare vs. Pclass
plt.figure(figsize=(8, 5))  # Mengatur ukuran figure
sns.scatterplot(x='Fare', y='Pclass', data=titanic_df, hue='Survived',
                palette="coolwarm",  # Mengatur palet warna
                markers=["o", "s"],  # Mengatur bentuk marker
                style="Sex")  # Mengatur style berdasarkan Sex
plt.title('Relationship between Fare, Pclass, and Survival', fontsize=14)  # Mengatur judul dan ukuran font
plt.xlabel('Fare', fontsize=10)  # Mengatur label sumbu x dan ukuran font
plt.ylabel('Pclass', fontsize=10)  # Mengatur label sumbu y dan ukuran font
plt.yticks([1, 2, 3])  # Mengatur ticks pada sumbu y
plt.legend(title="Survived", loc="best")  # Mengatur posisi dan judul legend
plt.show()
```

## Pertemuan 12: Clean a real-world dataset (from Kaggle or another source), perform normalization, handle outliers, and prepare the data for analysis.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# 1. Data Cleaning
# a. Handling Missing Values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# b. Dropping irrelevant columns
df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True)

# 2. Data Normalization
# Selecting numerical features for normalization
numerical_features = ['Age', 'Fare']

# Creating a MinMaxScaler object
scaler = MinMaxScaler()

# Fitting and transforming the selected features
df[numerical_features] = scaler.fit_transform(df[numerical_features])

# 3. Handling Outliers
# Using IQR method to detect and handle outliers for 'Age' and 'Fare'
for feature in numerical_features:
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Replacing outliers with upper and lower bounds
    df[feature] = np.where(df[feature] > upper_bound, upper_bound, df[feature])
    df[feature] = np.where(df[feature] < lower_bound, lower_bound, df[feature])

# 4. Data Preparation for Analysis
# a. Converting categorical features to numerical using one-hot encoding
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

print("Cleaned and prepared data:\n", df.head())
```
